# subplot

```python
x = [1,2,3,4,5,6]
y = [5,15,20,25,30,35]
y1 = [i*5 for i in x]
y2 = [i ** 2 for i in x]
y3 = [i+5 for i in x]
print(x)
print(y)
print(y1)
print(y2)
print(y3)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,ls="-")

plt.subplot(2,2,2)
plt.plot(x,y1,ls="--")

plt.subplot(2,2,3)
plt.plot(x,y2,ls=":")

plt.subplot(2,2,4)
plt.plot(x,y3,ls="-.")

plt.show()


```

#  Q2 analysis
## 1.arrange all elements in ascending order
## check the length of the data, if it is odd take middle element as median and if it is even pick middle two value and find average
## of two values as median

# Q1 internal mechanism

```python
q1 = np.percentile(student_scores,25)
print(q1)
q2 = np.percentile(student_scores,50)
print(q2)
q3 = np.percentile(student_scores,75)
print(q3)
```

```python
# position = (p/100) * (N - 1)
## p --> percentile
## N --> number of values
# position=(25/100)*(7-1)
# print(position) #1.5
# q1 = min_index_position_value+0.5*(max_index_position_value - min_index_position_value)
q1 = 80 + 0.5*(85-80)
print(q1)
```

```python
position_q3 = (75/100)*(6)
print(position_q3)
```

```python
q3 = 95 + 0.5*(105-95)
print(q3)
```

```python
# iqr(inter quartile range)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
lower_bound_liers = q1 - 1.5*iqr
print(lower_bound_liers)
```

```python
upper_bound_liers = q3 +1.5*iqr
print(upper_bound_liers)
```
