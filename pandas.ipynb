# concatenation

```python
dict2 = {"name": ["simon","leo","ranidher"],
        "age": [25,26,27]}
df2 = pd.DataFrame(dict2)
df2
```

# merge

```python
data1 = {"id":[1,2,3,4,5,6],
        "name":["den","rome","luck","vignesh","peo","ganesh"]}
data2 = {"id":[6,7,8,9,10,11],
        "place":["Banglore","hyd","vijayawada","Lucknow","andhra","up"]}
```

```python
df3 = pd.DataFrame(data1)
df4 = pd.DataFrame(data2)
```

```python
df3
```

```python
df4
```

```python
merged_df = pd.merge(df3,df4,on='id')
merged_df
```

```python
merged_df = pd.merge(df3,df4,how='outer')
merged_df
```

```python
merged_df = pd.merge(df3,df4,on="id",how='left')
merged_df
```

```python
merged_df = pd.merge(df3,df4,on="id",how='right')
merged_df
```

```python
merged_df3 = pd.merge(df3,df4,on='id',how="outer")
merged_df3
```

```python
merged_df4 = pd.merge(df3,df4,on='id',how="outer")
merged_df4
```

```python
data3 = {"emp_id":[1,2,3,4,5,6],
        "name":["luck","bad","good","ganesh","parmesh","ramesh"],
        "dept":["HR","IT","IT","Finance","Finance","HR"]}

data4 = {"emp_id":[1,2,3,4,5,6],
        "dept":["HR","IT","Finance","Finance","HR","IT"],
         "Place":["Bangalore","hyd","lucknow","pune","mangalore","tirupati"]}
```

```python
df5 = pd.DataFrame(data3)
df6 = pd.DataFrame(data4)
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6,on=["emp_id","dept"],how="inner")
```

```python
pd.merge(df5,df6,on=["emp_id","dept"],how="outer")
```
